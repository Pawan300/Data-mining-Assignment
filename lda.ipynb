```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer as wnl
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
punct_num = '''!()-àÂ[]{};:\n,<>./?@#%^"&\*_~0123456789=+|'"'''
def punctuation(x):
    no_punct = ""
    for char in x:
            if char not in punct_num :
                no_punct = no_punct + char  
    return(no_punct)
documents['headline_text']=documents['headline_text'].apply(lambda x:punctuation(x))
```

```python
stop=stopwords.words("english")
ps=PorterStemmer()
def preprocess(news):
        token=word_tokenize(news)
        token=[i for i in token if(i not in stop)]
        token=[wnl().lemmatize(i) for i in token]
        token=[ps.stem(i) for i in token]
        return(token)
```

```python
clean=documents['headline_text'].apply(lambda x:preprocess(x))
```
