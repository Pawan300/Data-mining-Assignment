```python
import numpy as np
import pandas as pd
import random
from sklearn.datasets import load_iris
```

```python
data=load_iris()
d=data["data"]
d.shape
```

```python
def initialize(d,k):
    a=np.random.randint(0,d.shape[0],k)
    seed=[]
    for i in a:
         seed.append(d[i])
    return(seed,list(a))
```

```python
def distance(seed,d):
    dist=np.zeros([d.shape[0],len(seed)])
    for i in range(d.shape[0]):
        for j in range(len(seed)):
            dist[i,j]=np.linalg.norm(d[i]-seed[j])
    return(dist)
```

```python
def kmediod(d,k):
    seed,seed_point=initialize(d,k)
    cluster=[""]*k
    dist=distance(seed,d)
    before=[]
    for i in dist:
        before.append(min(i))
    flag=0
    m=0
    while( flag==0):
        score=[]
        for i in range(k):
            for l in range(k):
                 cluster[l]=[]
        while(True):
            a=np.random.randint(0,d.shape[0])
            if(a not in seed_point):
                seed_point.append(a)
                break
        seed[i]=d[a]
        dist=distance(seed,d)
        for j in range(dist.shape[0]):
            cluster[np.argmin(dist[j])].append(d[j])
        after=[]
        for i in dist:
             after.append(min(i))
        score.append(sum(list(np.array(before) - np.array(after))))
        if(all(x<0 for x in score)):
            flag=1     
    return(cluster)
```

```python
cluster=kmediod(d,5)
```
