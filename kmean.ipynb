```python
import numpy as np
import pandas as pd
import random
import copy
from sklearn.datasets import load_iris
```

```python
data=load_iris()
d=data["data"]
```

```python
def initial_seed(d,k):
    a=random.randint(1,d.shape[0])
    cluster_point=[""]*k
    dist=-1
    max_dist=0
    temp=0
    for i in range(k):
        if(i==0):
            for j in d:
                di=np.linalg.norm(j-d[a])
                if di>dist:
                    dist=di
                    cluster_point[i]=j
                    temp=j
        else:
            di=[0]*i
            d=np.delete(d,temp,axis=0)
            for j in range(d.shape[0]):
                dist=di
                flag=0
                for m in range(i):
                    dist[m]=np.linalg.norm(j-cluster_point[m])
                for l in range(i):
                    if(di[l]!=dist[l]):
                        flag=1
                        break
                if(flag==0):
                    di=dist
                    cluster_point[i]=d[j]
                    temp=j
    return(cluster_point)
```

```python
def distance(d,cluster_point):
    dist=np.zeros([d.shape[0],len(cluster_point)])
    for i in range(d.shape[0]):
        for j in range(len(cluster_point)):
            dist[i,j]=np.linalg.norm(d[i]-cluster_point[j])
    return(dist)
```

```python
def kmean(d,k):
    cluster_point=initial_seed(d,k)
    cluster=[""]*k
    for i in range(k):
        for l in range(k):
            cluster[l]=[]
        dist=distance(d,cluster_point)
        for j in range(dist.shape[0]):
            cluster[np.argmin(dist[j])].append(d[j])
        for l in range(k):
            cluster_point[l]=np.mean(np.array(cluster[l]),axis=0)
    return(cluster)
```

```python
cluster=kmean(d,5)
```
